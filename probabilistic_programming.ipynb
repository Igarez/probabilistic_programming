# Garbage in Garbage Out

**If the way of thinking or the data we enter is incorrect, the solution will also be incorrect. These are some of the most difficult bugs to find since it is not a software failure.**



A common story about sampling is the case of the Allied planes in World War II. What happened was that they saw that the returning fighter planes usually had a lot of damage and impact of projectiles on the wings and other areas. What the RAF thought was to count the holes, make an average of their planes and start sending the reinforced planes in those areas. The problem? Their sample was planes that did return from combat, so Abraham Wald made them see that if the planes returned with holes in those parts, it was because in those parts the impacts were not lethal and contrary to what the data said (there are more holes in these points, so we must reinforce them), the other parts of the plane had to be reinforced, because those were the ones that did not return from combat

# Prejudice in sampling


* For a sampling to serve as a basis for statistical inference it must be random and representative
* Prejudice in sampling eliminates the representativeness of the samples
* Sometimes it is difficult to get samples, so the population with the easiest access is used (e.g. university studies)





It's like those Twitter surveys where all the people of X ideology share as much as possible, and in the end those who answered the survey were a homogeneous group (precisely those who adhere to those ideas)

And then, that survey is used as an argument to validate their belief, it's like a circle of self rejoicing and validation

Another example would be the same found in a newspaper of extreme left or extreme right is very likely that the results are very different but not representative because there is a prejudice in the sample.

# Texas sharpshooter fallacy




* This fallacy occurs when randomness is not taken into consideration
* It also happens when one focuses on the similarities and ignores the differences
* When we fail to have a hypothesis before collecting data we are at high risk of falling into this fallacy (Very common in data science)





is a reasoning that ignores all suggestive signs that an idea is wrong, in order to emphasize that information that seems to support it. Sometimes reality is deformed to this end, interpreting it in a disfigured way to approximate what is intended.

# Regression fallacy



* Many events fluctuate naturally, e.g. the average temperature of a city, the performance of an athlete, the returns of an investment portfolio, etc.

# Introduction to Machine Learning

It is the field of study that gives computers the ability to learn without being explicitly programmed.

**History of Machine Learning**

* It all starts with Bayes' Theorem
* In 1950 Alan Turing establishes the basis of computer learning (Computing machinery and intelligence).
* In 1951 Marvin Minsky creates the first neural network.
* In 1952 Arthur Samuel creates the first program that plays Chinese checkers, although it was a program that calculated the next moves, thanks to this we deduce that human learning is based on memorizing and generalizing.
* In 1957 Frank Rosenblat invents the perceptron that recognizes handwritten digits and represents them computationally.
* In 1963 the first adversarial network is programmed (it puts two computers in competition against each other and thus automates the rules of a game).
* In 1967, the nearest neighbours algorithm was invented, allowing the identification of patterns within existing data. It was applied to the calculation of the most efficient route.
* In 1969 Marvin Minsky writes the book Perceptrons, where he mathematically demonstrates the impossibility of the advance of Artificial Intelligence through neural networks (however he did not consider the possibility of adding layers of complexity to these), this stopped the research of neural networks for a time.
* In 1979 the Stanford cart was created, which was an autonomous indoor navigation vehicle.
* In 1992 the first program to play backgammon was developed.
* In 1997 IBM's Deep Blue defeated world chess champion Gary Kasparov.
* In 1998 the MNIST data set is released, consisting of hand-crafted digits and their respective labels.
* In 2009 Fei Fei Li released ImageNet, a dataset of quality images properly labeled.
* In 2011 IBM's Watson succeeds in winning the Jeopardy game.
* In 2012 a process of image identification from the YouTube dataset is implemented
* In 2014 Facebook implements a face recognition algorithm
* In 2016 Google's Alpha Go defeats Go world champion Fan Hui

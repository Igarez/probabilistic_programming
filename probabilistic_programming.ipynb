# Garbage in Garbage Out

**If the way of thinking or the data we enter is incorrect, the solution will also be incorrect. These are some of the most difficult bugs to find since it is not a software failure.**



A common story about sampling is the case of the Allied planes in World War II. What happened was that they saw that the returning fighter planes usually had a lot of damage and impact of projectiles on the wings and other areas. What the RAF thought was to count the holes, make an average of their planes and start sending the reinforced planes in those areas. The problem? Their sample was planes that did return from combat, so Abraham Wald made them see that if the planes returned with holes in those parts, it was because in those parts the impacts were not lethal and contrary to what the data said (there are more holes in these points, so we must reinforce them), the other parts of the plane had to be reinforced, because those were the ones that did not return from combat

# Garbage in Garbage Out

**If the way of thinking or the data we enter is incorrect, the solution will also be incorrect. These are some of the most difficult bugs to find since it is not a software failure.**



A common story about sampling is the case of the Allied planes in World War II. What happened was that they saw that the returning fighter planes usually had a lot of damage and impact of projectiles on the wings and other areas. What the RAF thought was to count the holes, make an average of their planes and start sending the reinforced planes in those areas. The problem? Their sample was planes that did return from combat, so Abraham Wald made them see that if the planes returned with holes in those parts, it was because in those parts the impacts were not lethal and contrary to what the data said (there are more holes in these points, so we must reinforce them), the other parts of the plane had to be reinforced, because those were the ones that did not return from combat

# Prejudice in sampling


* For a sampling to serve as a basis for statistical inference it must be random and representative
* Prejudice in sampling eliminates the representativeness of the samples
* Sometimes it is difficult to get samples, so the population with the easiest access is used (e.g. university studies)





It's like those Twitter surveys where all the people of X ideology share as much as possible, and in the end those who answered the survey were a homogeneous group (precisely those who adhere to those ideas)

And then, that survey is used as an argument to validate their belief, it's like a circle of self rejoicing and validation

Another example would be the same found in a newspaper of extreme left or extreme right is very likely that the results are very different but not representative because there is a prejudice in the sample.

# Texas sharpshooter fallacy




* This fallacy occurs when randomness is not taken into consideration
* It also happens when one focuses on the similarities and ignores the differences
* When we fail to have a hypothesis before collecting data we are at high risk of falling into this fallacy (Very common in data science)
